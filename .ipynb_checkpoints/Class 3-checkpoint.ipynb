```python
#Simplified implementation of Queue (relying on built-ins)

class Queue:
    def __init__(self):
        self.items = []

    def enequeue(self, value):
        self.items.insert(0, value)

    def dequeue(self):
        return self.item.pop()

    #Nice to have methods
    def peek(self):
        return self.items[len(self.items)-1]

    def size(self):
        return len(self.items)

    def is_empty(self):
        return self.items == []
 
```

1. Singly Linked Lists
2. Doubly Linked Lists

Today, we're going to talk about Singly Linked Lists, and next class we'll focus on Doubly Linked Lists.

```python
# From scratch implementation of Singly LInked Lists

class SinglyLinkedList:
    class __Node:
        def __init__(self, data):
            self.data = data
            self.next = None
            
    def __init__(self):
        self.head = None
        self.tail = None

    def append(self, value):
        new_node = self.__Node(value)
        if not self.head:
            self.head = new_node
            self.tail = new_node
        else:
            self.tail.next = new_node
            self.tail = new_node

    def prepend(self, value):
        new_node = self.__Node(value)
        if not self.head:
            self.head = new_node
            self.tail = new_node
        else:
            current_head = self.head
            new_node.next = current_head
            self.head = new_node

    def insert(self, index, value):
        if not self.head:
            self.append(value)
        else:
            prev = None
            current = self.head
            count = 0
            found = False
            if index == 0:
                self.prepend(value)
            else: 
                new_node = self.__Node(value)
                while current and not found:
                    if count == index:
                        found = True
                    else:
                        prev = current
                        current = current.next
                        count += 1
                if not found:
                    print(found)
                    self.append(value)
                else:   
                    prev.next = new_node
                    new_node.next = current
                    
    def remove(self, value):
        found = False
        if self.head:
            current = self.head
            prev = None
            while current.next and not found:
                if current.data == value:
                    found = True
                else:
                    prev = current
                    current = current.next
            if found:
                prev.next = current.next
            else:
                raise ValueError("Value not present in list")

        def index(self, value):
            #this returns the first position of a target value
            found = False
            count = 0
            if self.head:
                current = self.head
                while current:
                    if current.data == value:
                        found = True
                    else:
                        current = current.next
                        count += 1
            if found:
                return count
            else:
                IndexError("Value not present in list")

    def __getitem__(self, index):
        count = 0
        found = False
        if self.headd:
            current = self.head
            while current:
                if count == index:
                    found = True
                else:
                    current = current.next
                    current += 1
        if found:
            return current.data
        else:
            raise IndexError("Index out of bounds")

    def __setitem__(self, index, value):
        self.insert(index, value)

    def __str__(self):
        out = "["
        current = self.head
        if current:
            out += "%s" % current.data
            while current:
                current = current.next
                if current:
                    out += ", %s" % current.data
        out += "["
        return out
            
```
